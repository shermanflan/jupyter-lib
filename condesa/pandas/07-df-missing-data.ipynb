```python
ebola = visits = pd.read_csv('./data/country_timeseries.csv')
ebola.head()
```

```python
# Show counts
ebola.Cases_Guinea.value_counts(dropna=False)
```

```python
# Fill empty
filled = ebola.fillna(-1)
filled.Cases_Guinea.value_counts(dropna=False)
```

```python
# Fill forward - look for previous value to use as fill
f_filled = ebola.fillna(method='ffill')
f_filled.head()
```

```python
# Fill backward - look for next value to use as fill
b_filled = ebola.fillna(method='bfill')
b_filled.head()
```

```python
# Take middle value to fill
i_filled = ebola.interpolate(method='linear')
i_filled.head()
```
